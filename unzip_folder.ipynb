## Descompactar arquivos zip das pastas de dados externos

```python
import zipfile
import os
```

```python
# Defina o caminho da pasta onde estão os arquivos zip
pasta_zip = r'C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\hospitalar_zip_2022'

# Defina o caminho da pasta onde os arquivos serão extraídos
pasta_extracao = r'C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\hospitalar_cons_22'
```

```python
# Verifique se a pasta de extração existe; se não, crie-a
if not os.path.exists(pasta_extracao):
    os.makedirs(pasta_extracao)
    print(f'Pasta de extração criada: {pasta_extracao}')
```

```python
# Verifique se a pasta zip existe
if not os.path.exists(pasta_zip):
    print(f'Pasta de arquivos zip não encontrada: {pasta_zip}')
else:
    # Percorra todos os arquivos na pasta zip
    arquivos_encontrados = False
    for arquivo in os.listdir(pasta_zip):
        if arquivo.endswith('.zip'):
            arquivos_encontrados = True
            caminho_arquivo_zip = os.path.join(pasta_zip, arquivo)
            try:
                with zipfile.ZipFile(caminho_arquivo_zip, 'r') as zip_ref:
                    zip_ref.extractall(pasta_extracao)
                    print(f'{arquivo} descompactado com sucesso em {pasta_extracao}')
            except zipfile.BadZipFile:
                print(f'Erro ao descompactar o arquivo: {arquivo}. O arquivo pode estar corrompido.')

    if not arquivos_encontrados:
        print('Nenhum arquivo .zip encontrado na pasta especificada.')
```
