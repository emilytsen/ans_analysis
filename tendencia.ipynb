```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_S832 = pd.read_csv(r'C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\data_joelho\S832_merged.csv', sep=';')
df_S832
```

# eventos por mês

```python
eventos_por_mes = df_S832['ANO_MES_EVENTO'].value_counts().sort_index()

# Criar o gráfico
plt.figure(figsize=(12, 6))
eventos_por_mes.plot(kind='bar')

# Customizar o gráfico
plt.title('Número de Eventos por Mês')
plt.xlabel('Mês')
plt.ylabel('Número de Eventos')
plt.xticks(rotation=45)

# Adicionar os valores em cima de cada barra
for i, v in enumerate(eventos_por_mes):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

# eventos por estado

```python
eventos_por_estado = df_S832['UF_PRESTADOR'].value_counts().sort_values(ascending=False)
# Criar o gráfico
plt.figure(figsize=(12, 6))
eventos_por_estado.plot(kind='bar', color='orange')

# Customizar o gráfico
plt.title('Número de Eventos por Estado')
plt.xlabel('Mês')
plt.ylabel('Número de Eventos')
plt.xticks(rotation=45)

# Adicionar os valores em cima de cada barra
for i, v in enumerate(eventos_por_estado):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

```python
eventos_por_faixa_etaria = df_S832['FAIXA_ETARIA'].value_counts().sort_index()

# Lista das faixas etárias em ordem crescente
faixas_etarias_ordenadas = [
    '1 a 4', '5 a 9', '10 a 14', '15 a 19', 
    '20 a 29', '30 a 39', '40 a 49', '50 a 59', 
    '60 a 69', '70 a 79', '80 ou mais', 'Não identificado'
]

# Reordenar a Series conforme a lista ordenada
eventos_por_faixa_etaria = eventos_por_faixa_etaria.reindex(faixas_etarias_ordenadas)

# Criar o gráfico
plt.figure(figsize=(12, 6))
eventos_por_faixa_etaria.plot(kind='bar', color='purple')

# Customizar o gráfico
plt.title('Número de Eventos por Estado')
plt.xlabel('Faixa Etária')
plt.ylabel('Número de Eventos')
plt.xticks(rotation=45)

# Adicionar os valores em cima de cada barra
for i, v in enumerate(eventos_por_faixa_etaria):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
