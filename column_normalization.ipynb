```python
import pandas as pd
import os

# Caminho da pasta onde estão os arquivos CSV
pasta_csv = r'C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\hospitalar_det_22'

# Lista todos os arquivos CSV na pasta
arquivos_csv = [f for f in os.listdir(pasta_csv) if f.endswith('.csv')]

# Dicionário para armazenar o número de colunas na primeira linha de cada arquivo
colunas_por_arquivo = {}

for arquivo in arquivos_csv:
    caminho_arquivo = os.path.join(pasta_csv, arquivo)
    
    # Tenta ler a primeira linha do arquivo CSV
    try:
        df = pd.read_csv(caminho_arquivo, delimiter=',', nrows=1, header=None)
        num_colunas = len(df.columns)
        colunas_por_arquivo[arquivo] = num_colunas
    
    except Exception as e:
        print(f"Erro ao ler o arquivo {arquivo}: {e}")

# Exibir o número de colunas na primeira linha de cada arquivo
print("Número de colunas na primeira linha de cada arquivo CSV:")
for arquivo, num_colunas in colunas_por_arquivo.items():
    print(f"{arquivo}: {num_colunas} colunas")
```

```python
import os
import pandas as pd

# Caminho do diretório dos arquivos CSV
caminho_diretorio = r'C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\hospitalar_det_22'

# Lista para armazenar os DataFrames
dataframes = []

# Itera sobre todos os arquivos no diretório
for arquivo in os.listdir(caminho_diretorio):
    if arquivo.endswith('.csv'):
        caminho_arquivo = os.path.join(caminho_diretorio, arquivo)
        try:
            # Ler o arquivo CSV
            df = pd.read_csv(caminho_arquivo)
            
            # Adicionar as colunas teste1 e teste2 ao cabeçalho, mas sem valores
            df['teste1'] = pd.NA
            df['teste2'] = pd.NA
            
            # Adicionar o DataFrame à lista
            dataframes.append(df)
        except Exception as e:
            print(f'Erro ao ler o arquivo {arquivo}: {e}')

# Concatenar todos os DataFrames em um único DataFrame
df_completo = pd.concat(dataframes, ignore_index=True)

# Salvar o DataFrame combinado em um novo arquivo CSV
df_completo.to_csv(r'C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\hospitalar_det_22_completo.csv', index=False)
```
