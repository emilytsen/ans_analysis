### Web Crawler que coleta dados hospitalares de pacientes da ANS

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import os
```

```python
# Inicialize o serviço do ChromeDriver
service = Service()
```

```python
# Configure as opções do Chrome
options = Options()
# Define o diretório de download
download_dir = r'C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\cons_2022'
prefs = {"download.default_directory": download_dir}
options.add_experimental_option("prefs", prefs)
```

```python
# Inicie o ChromeDriver com o serviço e as opções configuradas
driver = webdriver.Chrome(service=service, options=options)
```

```python
# Abra a URL principal
url = 'https://dadosabertos.ans.gov.br/FTP/PDA/TISS/HOSPITALAR/2022/'
driver.get(url)

driver.implicitly_wait(10)

# Encontre todos os links para os estados
state_links = driver.find_elements(By.XPATH, "//table//tr/td/a[contains(@href, '/')]")

# Iterar sobre cada estado
for state_link in state_links:
    state_name = state_link.text.strip('/')
    print(f"Navegando para o estado: {state_name}")
    
    # Clicar no link do estado
    state_link.click()
    
    # Aguarde para garantir que a página está carregada
    driver.implicitly_wait(10)
    
    # Encontre todos os links para arquivos .zip
    zip_links = driver.find_elements(By.XPATH, "//table//tr/td/a[contains(@href, '.zip')]")
    
    # Iterar sobre cada arquivo .zip e clicar para download
    for zip_link in zip_links:
        file_name = zip_link.text
        print(f"Baixando arquivo: {file_name}")
        zip_link.click()
        time.sleep(2)  # Aguarde um pouco para o download começar
    
    # Voltar para a página principal
    driver.back()
    driver.implicitly_wait(10)
    
    # Recarregar os links dos estados
    state_links = driver.find_elements(By.XPATH, "//table//tr/td/a[contains(@href, '/')]")

# Fechar o navegador
driver.quit()
```
