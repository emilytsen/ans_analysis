## Separar arquivos zip de pastas pelo nome

```python
import os
import shutil #operações de alto nível para manipulação de arquivos e diretórios
```

```python
# Diretório onde estão os arquivos CSV
source_dir = r"C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\hospitalar_cons_22"

# Diretório de destino para os arquivos que contêm "DET"
destination_dir = r"C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\hospitalar_det_22"
```

```python
# Itera sobre os arquivos na pasta original
for file_name in os.listdir(source_dir):
    if "DET" in file_name and file_name.endswith(".csv"):
        # Caminho completo do arquivo original
        source_file = os.path.join(source_dir, file_name)
        
        # Caminho completo do destino
        destination_file = os.path.join(destination_dir, file_name)
        
        # Move o arquivo para a pasta de destino
        shutil.move(source_file, destination_file)
        print(f"Movido: {file_name}")
```
