## Contar numero total de linhas dos arquivos csv


```python
def count_rows(folder_path):
    total_rows = 0
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path, delimiter=';')
            total_rows += len(df)
    return total_rows

# Especifica o caminho da pasta
folder_path = r'C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\hospitalar_det_22'

# Chama a função e imprime o resultado
total_linhas = count_rows(folder_path)
print(f'Total de linhas somadas de todos os arquivos CSV: {total_linhas}')
```

```python
import os
import pandas as pd

# Define o caminho da pasta que contém os arquivos CSV
folder_path = r'C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\hospitalar_det_22'

# Cria uma lista para armazenar os DataFrames individuais
dfs = []

# Itera sobre todos os arquivos na pasta
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        
        # Lê o arquivo CSV
        df = pd.read_csv(file_path, delimiter=';', low_memory=False)
        
        # Adiciona o DataFrame à lista
        dfs.append(df)

# Concatena todos os DataFrames da lista em um único DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Salva o DataFrame combinado em um novo arquivo CSV
combined_df.to_csv(r'C:\Users\etsen\OneDrive - JACTO\Síntegra Digital Health\Data Science\DadosExternos\ans\beneficiarios\hospitalar_2022\hospitalar_det_2022.csv', index=False, sep=';')

print("Todos os arquivos foram combinados em um só.")
```
